```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1. Loading the data file using Pandas
df = pd.read_excel("restaurant_data.xlsx")
df.head()
```

```python
cntry_code = pd.read_excel("Country-Code.xlsx")
cntry_code.head()
```

```python
#Using merge function to merge the 2 files using Country Code
df_new = pd.merge(df, cntry_code,on='Country Code', how='left')
df_new.head()
```

```python
# Addinng profer format for col names and getting column info
df_new.columns = df_new.columns.str.replace(' ', '_')
df_new.columns
```

```python
#Description of columns
df_new.info()
```

```python
# 2. Check for null values in the data. 
# Get the number of null values for each column
df_new.isnull().sum()
```

```python
df_new[df_new['Restaurant_Name'].isnull()]
```

```python
#Dropping the row with no restaurant_name and resetting index
df_new.dropna(axis=0, subset=['Restaurant_Name'], inplace=True)
df_new.reset_index(drop=True, inplace=True)
```

```python
df_new[df_new['Cuisines'].isnull()]
```

```python
#Marking records with Cusisines as NULL to 'Others'
df_new['Cuisines'].fillna('Others', inplace=True)
```

```python
#checking for nulls and returning col info
df_new.isnull().sum
df_new.info()
```

```python
# Saving the changes to excel file to use in Tableau for Dashboard creation
#Creating exel writer object
writer = pd.ExcelWriter('restaurant_output.xlsx')

#write dataframe to excel
df_new.to_excel(writer)

#save the excel
writer.save()
```

```python
# Performing EDA
# EDA 1. Identifying count of restaurants by countries
ctry_res_count = df_new.groupby(['Country_Code','Country']).agg( Count= ('Restaurant_ID','count'))
ctry_res_count.sort_values(by='Count', ascending=False)
```

```python
ctry_res_count.plot(kind='bar')
```

```python
# getting the restaurant count by city using group by on Country
city_res_count = df_new.groupby(['Country','City']).agg( Count= ('Restaurant_ID','count'))
city_res_count.sort_values(by='Count', ascending=False)
```

```python
city_res_count.describe()
#The data seems to be skewed towards New Delhi, Gurgaon and Noida. There is minimal data for other countries/cities
# New Delhi has max restaurant has count = 5473
# city with min restaurant has count = 1
```

```python
min_res_cnt = city_res_count[city_res_count['Count']==1]
min_res_cnt                          
#  46 cities in different countries have 1 restaurant
```

```python
# EDA 2: Explore the franchise with most national presence
# Since majority count of restaurants is from India I am creating a new data set to verify the stats for India

df_India = df_new.loc[df_new['Country'] == 'India']
df_India.head()
```

```python
#Identifying restauranct franchise with national presence

res_count_India = df_India.groupby(['Restaurant_Name']).agg( Count= ('Restaurant_ID','count'))
res_count_India.sort_values(by='Count', ascending=False)

# Cafe Coffee Day has most restaurants in India
```

```python
# Exploring data for Cafe Coffee Day since it has most national presence
#Average cost for 2 at Cafe Coffee Day restaurant (In Indian rupees)

avg_cost_for_2 = df_India[df_India['Restaurant_Name']=='Cafe Coffee Day']['Average_Cost_for_two'].value_counts()
avg_cost_for_2     
#Average cost for 2 people to eat at Coffe Cafe Day is 450 rupees which is pretty reasonable price
```

```python
#Rating text for top franchise

ratings_for_top_fran = df_India[df_India['Restaurant_Name']=='Cafe Coffee Day']['Rating_text'].value_counts()
ratings_for_top_fran
# Most of the restaurants for the franchise are rated Average, 16 are not rated
```

```python
#Checking Price ranges for the franchise

price_range_for_fran = df_India[df_India['Restaurant_Name']=='Cafe Coffee Day']['Price_range'].value_counts()
price_range_for_fran.sort_values(ascending=False).head()

#All the restaurants under this franchise fall in price range 1
```

```python
#Aggregate ratings for this franchise

agg_rate_for_top_fran = df_India[df_India['Restaurant_Name']=='Cafe Coffee Day']['Aggregate_rating'].value_counts()
agg_rate_for_top_fran

# There is no agg_rating for 16 restaurants for Cafe Coffee Day franchise
# Highest rating is 3.6 and 2 of them have this
```

```python
# EDA 3: Ratio between restaurants that allow table booking vs. those that do not allow table booking

booking_yes = df_India['Has_Table_booking'].value_counts()['Yes']
booking_no = df_India['Has_Table_booking'].value_counts()['No']
print('Ratio between restaurants that allow table booking vs. those that do not allow table booking: ',
      round((booking_yes/booking_no),2))

# Ratio is 0.15, means lot of the resturants doesn't allow table booking
```

```python
#Pie chart showing percentage difference for table booking Yes/No

plt.figure(figsize=(10,8))
x = df_India.Has_Table_booking.value_counts()
y = df_India['Has_Table_booking'].value_counts().index
plt.pie(x, labels=y, autopct='%.2f')
plt.title('Table Booking')
plt.show()
```

```python
# EDA 4: Percentage of restaurants providing online delivery
online_delivery_yes =df_India['Has_Online_delivery'].value_counts()['Yes']
print('Percentage of restaurants providing online delivery : {} %'.format((round(online_delivery_yes/len(df_India),2)*100)))

# 28% percent of restaurants allow online delivery
```

```python
#Pie chart showing online delivery yes/no

plt.figure(figsize=(10,8))
x = df_India.Has_Online_delivery.value_counts()
y = df_India['Has_Online_delivery'].value_counts().index
plt.pie(x, labels=y, autopct='%.2f')
plt.title('Online Delivery')
plt.show()
```

```python
# EDA 5: Difference in number of votes for the restaurants that deliver and the restaurants that do not deliver

res_deliver = df_India[df_India['Has_Online_delivery'] == "Yes"]['Votes'].sum()
print(res_deliver)
res_no_deliver =  df_India[df_India['Has_Online_delivery'] == "No"]['Votes'].sum()
print(res_no_deliver)
print('Difference in number of votes for restaurants that deliver and dont deliver: ',abs((res_deliver - res_no_deliver)))

#Difference in number of votes is coming to 173166
```

```python
# EDA 6: Top 10 cuisines served across cities
# What is the maximum and minimum number of cuisines that a restaurant serves?
# Which is the most served cuisine across the restaurant for each city?

#Cuisine has multiple entries combined hence splitting them and adding the entries to list variable called total_cuisines
total_cuisines=[]
k = df_India[df_India['Cuisines'].notnull()]
k['Cuisines'] = k['Cuisines'].apply(lambda x:x.strip())
for i in k['Cuisines']:
    for j in i.split(','):
        j = j.strip()
        total_cuisines.append(j)    
```

```python
# Getting top 10 cuisines from the list
top_10_cuisines=pd.Series(total_cuisines).value_counts()[:10]
top_10_cuisines

# North Indian, Chinese, Fast Food, Mughlai, Bakery, Continental, Italian, South Indian, Cafe and Desserts are top 10 cuisines
```

```python
plt.figure(figsize=(10,5)) 
a=pd.Series(total_cuisines).value_counts()[:10]
a.plot(kind='bar')
plt.title('Top 10 cuisines by restaurants')
plt.xlabel('Cuisine type')
plt.ylabel('Number of occurances')
```

```python
# Which is the most served cuisine across the restaurant for each city?

#Creating data frame for restaurants with the cuisine list

df_2 = df_India[df_India['Cuisines'].isin(total_cuisines)]
df_2
```

```python
# Which is the most served cuisine across the restaurant for each city?

most_serv =pd.DataFrame(df_2.groupby(by=['Restaurant_Name','City','Cuisines']).size().reset_index()) 

most_serv.columns=['Restaurant_Name','City','Cuisines','Count']
most_serv.sort_values('Count',ascending=False).head()

# Cafe Coffee Day in New Delhi has most cuisines served which is 57
# Least number Cuisine served is 1
```
